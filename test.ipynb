```python
file_name="Seating Arrangement for B Tech Semester VI-(MECH)_Department Elective Courses_17-05-2024 and 18-05-2024.pdf"
```

```python
import fitz  # PyMuPDF

def extract_text_from_non_ocr_pdf(pdf_path):
    # Open the PDF file
    document = fitz.open(pdf_path)

    # Initialize an empty string to store extracted text
    extracted_text = ""

    # Iterate through each page
    for page_num in range(len(document)):
        page = document.load_page(page_num)
        
        # Use OCR to extract text from the page
        text = page.get_text("text")
        
        # Append extracted text to the result
        extracted_text += text

    # Close the document
    document.close()

    return extracted_text
```

```python
# Replace with your PDF file path
pdf_path = file_name

# Extract text from non-OCR PDF
extracted_text = extract_text_from_non_ocr_pdf(pdf_path)

# Print the extracted text
print(extracted_text)
```
